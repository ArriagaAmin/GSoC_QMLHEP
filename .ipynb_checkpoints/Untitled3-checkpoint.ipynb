```python
import numpy as np

b = np.load('QIS_EXAM_200Events.npz')
trainingData = b['training_input'].tolist()

trainingData = np.append(trainingData['0'], trainingData['1'], axis = 0)

expected = [0 for i in range(50)]
for i in range(50): expected.append(1)
expected = np.array(expected).T

def sigmoid(x, deriv=False):
    if deriv:
        return x*(1-x)
    return 1/(1+np.exp(-x))

np.random.seed(0)

## Creamos la red
DIM = 100
SYN0 = 2*np.random.random((5,DIM)) - 1
SYN1 = 2*np.random.random((DIM,DIM)) - 1
SYN2 = 2*np.random.random((DIM, 1)) - 1

for i in range(20000):
    
    # FORWARD
    l0 = trainingData
    l1 = sigmoid(np.dot(l0, SYN0))
    l2 = sigmoid(np.dot(l1, SYN1))
    l3 = sigmoid(np.dot(l2, SYN2))
 
    # ERROR
    l3_error = expected - l3
    # CORRECTION
    l3_delta = l3_error * sigmoid(l3, True)
 
    l2_error = np.dot(SYN2.T, l3_delta)
    l2_delta = l2_error * sigmoid(l2, True)   
    
    l1_error = np.dot(l2_delta, SYN1.T)
    l1_delta = l1_error * sigmoid(l1, True)
 
    ## Las primeras conexiones las corregimos
    ## con l1_delta * los inputs
    SYN0 += np.dot(l0.T, l1_delta)
    ## Las segundas conexiones las corregimos con
    ## l2_delta * los inputs de la capa intermedia
    SYN1 += np.dot(l1.T, l2_delta)
    
    #SYN2 += np.dot(l2.T, l3_delta)
        
    print("Error:" + str(np.mean(np.abs(l2_error))))
    
print(l2)
```
